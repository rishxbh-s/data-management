```python
import pandas as pd

df = pd.read_csv('feedbacks.csv')
df
```

```python
df = df.drop_duplicates(subset=['Full Name', 'Email Address'])
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(df['Rate your experience with the overall event'], bins=5, kde=False, color='skyblue')
plt.title('Distribution of Overall Event Ratings')
plt.xlabel('Ratings')
plt.ylabel('Frequency')
plt.show()
```

```python
activities = ['Rate the activity-1', 'Rate the following activity-2', 'Rate the following activity-3', 'Rate the following activity-4']
averages = df[activities].mean()

averages.plot(kind='bar', color='orange', edgecolor='black')
plt.title('Average Ratings by Activity')
plt.ylabel('Average Rating')
plt.xlabel('Activities')
plt.show()
```
