```python
import pandas as pd

faculty_data = pd.read_csv('teacher_details.csv') 
leaders_data = pd.read_csv('domain_leader.csv')  
helpers_data = pd.read_csv('domain_helpers.csv')  

domain_data = pd.concat([leaders_data, helpers_data]).drop_duplicates()

faculty_data['Valid'] = ~faculty_data['Registered Email ID'].isin(domain_data['Email'])


faculty_data['Associated'] = faculty_data['College Name'].isin(domain_data['College Name'])
faculty_data['Associated'] = faculty_data['Associated'].map({True: 'Associated', False: 'Not Associated'})

valid_faculty = faculty_data[(faculty_data['Valid']) & (faculty_data['Associated'] == 'Associated')]

valid_faculty.to_csv('valid_faculty.csv', index=False)
df = pd.read_csv('valid_faculty.csv')
df
```
