```python
import pandas as pd


feedback = pd.read_csv('feedbacks.csv')  
discord = pd.read_csv('discord_entries.csv')  
feedback['Responded'] = 'Yes'


combined = pd.merge(discord, feedback, how='left', on='Email Address')

combined['Responded'] = combined['Responded'].fillna('No')

combined.to_csv('consolidated_data.csv', index=False)
df= pd.read_csv('consolidated_data.csv')
df
```
